```python
import os
from openpyxl import load_workbook
from openpyxl.drawing.image import Image

# 加载Excel文件
excel_path = r'C:\Users\86155\Desktop\学习\新建文件夹\统计信息1.xlsx'  
wb = load_workbook(excel_path)
ws = wb.active

# 设置图片文件夹路径
folder_path = r'C:\Users\86155\Downloads\土库曼斯坦'  

image_column = 5

# 遍历文件夹中的PNG文件
for filename in os.listdir(folder_path):
    if filename.endswith('.png'):
        # 从文件名中去除扩展名，获得图片的基本名称
        base_name = os.path.splitext(filename)[0]

        # 在Excel表格中查找该名称所在的行
        for row in ws.iter_rows(min_row=2):  
            if row[1].value == base_name:  
                # 获取图片文件的路径
                img_path = os.path.join(folder_path, filename)

                # 创建图片对象
                img = Image(img_path)

                # 将图片插入Excel的指定单元格
                cell = f'{chr(64 + image_column)}{row[0].row}' 
                ws.add_image(img, cell)
                break

# 保存修改后的Excel文件
output = r'C:\Users\86155\Desktop\学习\新建文件夹\统计信息2.xlsx'
wb.save(output)
```
